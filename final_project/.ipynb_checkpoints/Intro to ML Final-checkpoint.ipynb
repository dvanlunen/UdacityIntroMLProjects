```python
import sys
import pickle
import pandas as pd
import numpy as np
from ggplot import *
from IPython.display import Image, display
from sklearn.naive_bayes import GaussianNB
from sklearn.cross_validation import train_test_split
from sklearn.decomposition import PCA
from sklearn import preprocessing
from sklearn.grid_search import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import os
import re
from sklearn.feature_extraction.text import TfidfVectorizer

from tester import dump_classifier_and_data
from poi_email_addresses import poiEmails
sys.path.append("../tools/")
from feature_format import featureFormat, targetFeatureSplit
from parse_out_email_text import parseOutText
```

```python
# Load numerical data
# Dict of Dicts
#    Higher level keys are LASTNAME FIRSTNAME MIDINITIAL
#    Lower level keys are features (e.g., bonus, salary)
with open("final_project_dataset.pkl", "r") as data_file:
    data_dict = pickle.load(data_file)
```

```python
# explore possible numerical features to include
pot_features = data_dict[data_dict.keys()[0]].keys()
print(pot_features)
print(data_dict.keys())
```

```python
# remove the 'TOTAL' key which is an outlier
data_dict.pop('TOTAL', None)
```

```python
# explore potential features
pot_features_df = pd.DataFrame.from_dict(data_dict, orient='index')
pot_features_df = pot_features_df.replace('NaN', np.nan)
pot_features_df.head()
```

```python
# See how much missing data there is for each feature
pot_features_df.isnull().sum().sort_values()
```

```python
# make histograms of all numeric columns
for feature in pot_features:
    if feature != 'email_address':
        p = ggplot(pot_features_df, aes(x=feature)) + geom_histogram()
        p.save('./histograms/{}.png'.format(feature))
        display(Image('./histograms/{}.png'.format(feature)))
```

```python
# Data Setup

# choose features
# messages with poi likely to have a big impact because 
#     those in conspiracy communicate with each other
# financial information added as well to explore
features_list = ['poi',
                 'from_this_person_to_poi', 'from_poi_to_this_person',
                 'salary', 'total_stock_value', 'restricted_stock', 'expenses', 
                 'bonus', 'long_term_incentive', 'deferred_income'
                ]
# convert information to a list of np arrays
# NaN converted to 0
data = featureFormat(data_dict, features_list, sort_keys=True)
labels, features = targetFeatureSplit(data)

# scale features for the classifiers that will be explored which need it
features_scaled = preprocessing.MinMaxScaler().fit_transform(features)

# split into test vs train
X_train, X_test, y_train, y_test = train_test_split(
features_scaled, labels, test_size=.1, random_state=8
)
# there are only 18 pois, will want to check pois are in both test and train
print(sum(y_train), sum(y_test))
```

```python
# try some dimension reduction using PCA
# use auto svd solver
# question: will the number of emails features make up different PCs than financials?
# answer: not really, seems like all the features are important in the first component and some less in the second
pca = PCA(n_components=2)
pca.fit(X_train)
print(pca.components_)
```

```python
# apply PCA to data
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# train an SVM with and without PCA
# search over a parameter grid for the best model
# use balanced to emphasize POIs

# PCA has no impact on classifications - likely because we started with so few variables in the first place
# However it does seem to impact the best values of C and gamma.
# with PCA: 
#     C is lower allowing a more smooth boundary now that we have fewer features
#     Gamma is higher allowing points further from the boundary to have more influence
param_grid = {
              'C': [1e1, 1e2, 1e3, 5e3, 1e4, 5e4, 1e5],
              'gamma': [0.00001, 0.0001, 0.0005, 0.001, 0.005, 0.01, 0.1],
             }
clf_svm = GridSearchCV(SVC(kernel='rbf', class_weight='balanced'), param_grid)
clf_svm.fit(X_train, y_train)
print('Best non-PCA SVM')
print clf_svm.best_estimator_
svm_pred = clf_svm.predict(X_test)
print confusion_matrix(y_test, svm_pred, labels=range(2))
print classification_report(y_test, svm_pred, target_names=['nonPOI', 'POI'])
print accuracy_score(svm_pred, y_test), "\n"

clf_svm.fit(X_train_pca, y_train)
print('Best PCA SVM')
print clf_svm.best_estimator_
svm_pred = clf_svm.predict(X_test_pca)
print confusion_matrix(y_test, svm_pred, labels=range(2))
print classification_report(y_test, svm_pred, target_names=['nonPOI', 'POI'])
print accuracy_score(svm_pred, y_test), "\n"
```

```python
# Parse emails
word_data = []
from_data = []
poi_emails = set(poiEmails())

# check if previously parsed so you can just access the pickle file
if os.path.isfile("word_data.pkl"):
    print("Loading Pre-Pickled Data")
    word_data = pickle.load(open('word_data.pkl', 'rb'))
    from_data = pickle.load(open('from_data.pkl', 'rb'))
# otherwise need to parse and save to pickle file
else:
    print("Loading Emails and Saving Pickle")
    for from_person in os.listdir('./emails_by_address'):
        from_person_locs = open('./emails_by_address/' + from_person, 'r')
        for path in from_person_locs:
            path = os.path.join('..', path[20:-1])
            try:
                email = open(path, 'r')
                text = parseOutText(email)
                word_data.append(text)
                if from_person[5:-4] in poi_emails:
                    from_data.append(1)
                else:
                    from_data.append(0)
                email.close()
            except IOError as e:
                print "I/O error({0}): {1}".format(e.errno, e.strerror)
            except:
                print "Unexpected error:", sys.exc_info()[0]
        from_person_locs.close()
    pickle.dump(word_data, open("word_data.pkl", 'w'))
    pickle.dump(from_data, open("from_data.pkl", 'w'))
```
